# 2. NumPy

## 2.1. Array creation

```python
import numpy as np

arr_1 = np.array([1, 2, 3, 4, 6])
print(arr_1)
print(type(arr_1))

arr_2 = np.arange(10)
print(arr_2)
print(type(arr_2))
```

### Evenly spaced numbers over a specified interval

```python
arr = np.linspace(0, 100, 10)
arr
```

## 2.2. Array reshaping

```python
arr = np.arange(100).reshape(10, 10)
print(arr)
arr.shape
```

```python
arr = np.arange(100).reshape(2, 5, 10)
print(arr)
arr.shape
```

## 2.3. Basic array operations

### Sum

```python
arr = np.arange(16).reshape(4, 4)
print(arr)

print(arr.sum()) # Sum over all array
print(arr.sum(axis=0)) # Sum over columns
print(arr.sum(axis=1)) # Sum over rows
```

### Min and Max

```python
print(arr.max())
print(arr.min())
print(arr.min(axis=1))
print(arr.min(axis=0))
```

### Arrays of ones and zeros

```python
arr = np.zeros(10)
arr
```

```python
arr = np.ones(10)
arr
```

```python
arr = np.ones(10)
arr.mean() # Array mean
```

```python
arr = np.linspace(0, 100, 10)
arr.std() # Array std
```

```python
arr_1 = np.array([1, 2, 3, 4, 6])
arr_2 = np.arange(5)

np.add(arr_1, arr_2) # Sum of two arrays
```

```python
arr_1 * 5 # Element wise multiplication
```

```python
arr_1 * np.pi
```

```python
arr_1 ** 5 # Element wise power
```

```python
value = 100
np.sqrt(value)
```

```python
arr_1 = np.array([1, 2, 3, 4, 6])
arr_2 = np.arange(5)

arr_1 * arr_2 # Element wise multiplication
```

```python
arr_1 = np.array([1, 2, 3, 4, 6])
arr_2 = np.arange(5)

np.multiply(arr_1, arr_2) # Element wise multiplication
```

```python
arr = [0, np.pi / 2, np.pi / 3, np.pi]
sin_Values = np.sin(arr)

print(f"Input array: {arr} \nSin Values: {sin_Values}")
```

## 2.4. Array manipulation

```python
arr_1 = np.array([1, 2, 3, 4, 6])
arr_2 = np.arange(5)

print(np.concatenate((arr_1, arr_2), axis=0))
print(np.vstack((arr_1, arr_2)))
print(np.hstack((arr_1, arr_2)))
print(np.column_stack((arr_1, arr_2)))
```

```python
arr = np.array([11, 2, 13, 4, 16])
print(arr)
print(np.sort(arr))
```

## 2.5. Boolean filtering

```python
arr = np.arange(25).reshape(5,5)
arr
```

```python
mask = arr >= 13
arr[mask]
```

```python
mask_2 = (arr > 13) & (arr < 22)
arr[mask_2]
```
